```python
from utils.utils import *
```

```python
x_path = '../../data/H264_raw/BasketballPass_416x240_50_H264_42_Ionly.yuv'
y_path = '../../data/YUV/BasketballPass_416x240_50.yuv'
```

```python
outputdir = '../../data/preprocessed/basketball/'
```

```python
height = int(y_path.split('_')[1].split('x')[1])
width = int(y_path.split('_')[1].split('x')[0])
size = (height, width)
```

```python
frames = range(500)
N_patches_per_frame = 100
size_of_patches = (64,64)
```

```python
X = read_YUV420_multiframes(x_path, size, frames, channel='Y')
Y = read_YUV420_multiframes(y_path, size, frames, channel='Y')
```

```python
ix = 0
iy = 0
for patch in range(N_patches_per_frame):
    idxs = n_random_slices(1, frames, size, size_of_patches)
    X_patches = patches_from_idxs(X, idxs)
    Y_patches = patches_from_idxs(Y, idxs)
    for ex in X_patches:
        np_to_pil(ex).save(outputdir+'ground_truth/'+str(ix)+'.png')
        ix += 1
    for ex in Y_patches:
        np_to_pil(ex).save(outputdir+'compressed/'+str(iy)+'.png')
        iy += 1
```
